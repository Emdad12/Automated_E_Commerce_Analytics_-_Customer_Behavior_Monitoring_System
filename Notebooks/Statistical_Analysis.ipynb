```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
import statistics
```

```python
data=pd.read_csv("Documents/ML_Competetion/Python Project/Sales project 1/data.csv")
data.head()
```

```python
med=data['revenue'].median()
data['is_high_spender'] = np.where(data['revenue'] >= med, 1, 0)
```

```python
data.info()
```

```python
from scipy.stats import skew,kurtosis,shapiro,mannwhitneyu,kruskal
```

```python
skew(data['revenue']),kurtosis(data['revenue'])
```

```python
data['log_revenue']=np.log1p(data['revenue'])
```

```python
skew(data['log_revenue']),kurtosis(data['log_revenue'])
```

```python
data[['log_revenue','promo_amount','shipment_fee','duration_days']].corr(method='pearson')

```

```python
data[['log_revenue','promo_amount','shipment_fee','duration_days']].corr(method='spearman')
```

```python
shapiro(data.sample(5000)['revenue'])
```

```python
promo=data[data['has_promo']==1]['revenue']
no_promo=data[data['has_promo']==0]['revenue']
mannwhitneyu(promo,no_promo,alternative='two-sided')
```

```python
repeat_customer=data[data['is_repeat_customer']==1]['revenue']
not_repeated=data[data['is_repeat_customer']==0]['revenue']
mannwhitneyu(repeat_customer,not_repeated,alternative='two-sided')
```

```python
groups = [
    g["revenue"].values
    for _, g in data.groupby("payment_method")
]

kruskal(*groups)
```

```python
promoo=data[data['has_promo']==1]['log_revenue']
no_promoo=data[data['has_promo']==0]['log_revenue']
def cohens_d(x,y):
    nx,ny=len(x),len(y)
    pooled_sd=np.sqrt(
        ((nx-1)*x.var()+(ny-1)*y.var())/(nx+ny-2)
    )
    return (x.mean()-y.mean())/pooled_sd
    
```

```python
cohens_d(promoo,no_promoo)
```

```python
import statsmodels.api as sm

x=data[['quantity','shipment_fee','is_repeat_customer','has_promo','is_high_spender']]
y=data['log_revenue']
X=sm.add_constant(x)
model=sm.OLS(y,X).fit()
model.summary()


```

```python
data['promo_repeat_interaction']=(data['has_promo']*data['is_repeat_customer'])
x=sm.add_constant(data[['quantity','shipment_fee','is_repeat_customer','has_promo','promo_repeat_interaction']])
y=data['log_revenue']
X=sm.add_constant(x)
model=sm.OLS(y,X).fit()
model.summary()
```

```python
from statsmodels.discrete.discrete_model import Logit
```

```python
x=sm.add_constant(data[['quantity','shipment_fee','is_repeat_customer','has_promo']].fillna(0))
y=data['is_high_spender']
logit=Logit(y,x).fit()
logit.summary()
```

```python
print(np.exp(logit.params))
```
